```python
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image
```

```python
bird_images = tf.keras.utils.image_dataset_from_directory(
    '../data/images',
    labels=None,
    label_mode=None,
    batch_size=64,
    image_size=(64,64),
    color_mode='grayscale'
)
```

```python
def preprocess_images(image):
    return (image - 127.5) / 127.5
```

```python
bird_images = bird_images.map(preprocess_images)
```

```python
for batch in bird_images:
    plt.imshow(batch[0], cmap='gray')
    break
```
