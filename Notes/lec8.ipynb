### Convolutional Neural Networks Pt. 2


##### Stride, Padding, Receptive field

Convention: 16@74x74 = d: 16, h: 74, w: 74

**Stride** = steps by kernel/filter

**Padding (zero-padding)**<br>
Increase width (zeroes) (Allows kernel/filter to cover edge pixels)
- Also useful to retain the size of the matrix
- Modes: valid, same, full




**Receptive field** 
- How much every feature/neuron is allowed to "see"

**Dilated convolution**<br>
An expanded kernel, not very common.
More of an example to show that kernels must look the same.
- Increase size of receptive field exponentially.

RGB channels = depth<br>
-> Still one feature map

One filter -> one feature map

---

Example from [Anodet](https://github.com/OpenAOI/anodet/blob/master/anodet/feature_extraction.py)

```
# Initial convolutional layer
batch = self.backbone.conv1(batch)  

# Batch normalization (BatchNorm2d)
batch = self.backbone.bn1(batch)  

# Rectified Linear Unit (ReLU) activation function
batch = self.backbone.relu(batch)  

# Downsampling to reduce spatial dimensions and retain features
batch = self.backbone.maxpool(batch)  

# Capture low-level features and spatial patterns
layer1 = self.backbone.layer1(batch)  

# Capturing more complex features
layer2 = self.backbone.layer2(layer1)

# Capture higher-level features
layer3 = self.backbone.layer3(layer2)

# Refine features, prepare for final classification
layer4 = self.backbone.layer4(layer3)
```
