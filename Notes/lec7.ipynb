## What is a convolution?

![](https://upload.wikimedia.org/wikipedia/commons/thumb/2/21/Comparison_convolution_correlation.svg/2560px-Comparison_convolution_correlation.svg.png)
from [wikipedia.org/wiki/Convolution](https://en.wikipedia.org/wiki/Convolution)

- Combinining two signals to form a third signal.

- Measure overlap of signal strength.

Signal = feature in the input data

- Convolutional operations
    - Apply convolutional operation on input
    - Moving a filter (or kernel) across the input data and computing the dot product between the filter and the local regions of the input.
- Signal strength
    - The filter/kernel contains weights that are learned during training. The dot product of filter and input measures the strength of the signal (presence of a feature in the local region).

- Overlap of signal strength
    - By moving the filter across the entire input, the convolutional operation captures spatial relationships (detects patterns), allowing the network to learn hierarchial representations of features.

- CNN = More efficient network
- convolution = cheaper than matmul
- Handle small changes (e.g., translations)
- Dynamic input sizes

### Convolutional layer

```
Input Image     x   Filter/Kernel   =  Feature Map (new signal)
[h x w matrix]      [3 x 3 matrix]         
                    size:     f=3
                    stride:   s=1
                    padding:  p=0
```

Backprop on the filter and its weights 

Padding = add extra border pixels to the input data (allow filter to start earlier) (adding zeroes)

No stride (stride = 1) = one pixel

If the stride is set to a value greater than 1 (e.g., stride = 2), the filter skips pixels as it moves across the input.

### Feature map

- Ouput from conv layer
- Find uniqueness (e.g., find horizontal lines, find circles)

Multiple kernels can be used together

### Max pooling & average pooling

- Pooling (often) used after conv layers
- "Zooming out"
- Dim reduction

- Max: only keep max value (most common), avg: only keep avg
