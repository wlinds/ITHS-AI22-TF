[World Economic Forum: Top 9 ethical issues in artificial intelligence](https://www.weforum.org/agenda/2016/10/top-10-ethical-issues-in-artificial-intelligence/)

1. Unemployment.
    - What happens after the end of jobs?
2. Inequality.
    - How do we distribute the wealth created by machines?
3. Humanity.
    - How do machines affect our behaviour and interaction?
4. Artificial stupidity.
    - How can we guard against mistakes?
5. Racist robots.
    - How do we eliminate AI bias?
6. Security.
    - How do we keep AI safe from adversaries?
7. Evil genies.
    - How do we protect against unintended consequences?
8. Singularity.
    - How do we stay in control of a complex intelligent system?
9. Robot rights.
    - How do we define the humane treatment of AI?

### Checklist for AI Projects

- What are we trying to solve?
- Are we introducing any bias?
    - Evaluate
- Why are we doing this?
- How is the end-user affected?
- Put the person in the center.
    - Don't let a model decide everything.
- Continous evaluation
    - What worked yesterday might not work today.
- Include diversity in the process
- Rollback if something goes wrong
