# Gradient Descent Simulation Exercise
### Objective
Understand the principles of gradient descent by manually computing and applying gradient descent steps to a simple function.

### Background
Gradient descent is an optimization algorithm used to minimize a function by iteratively moving towards the minimum value of the function.

### Simple Function
We'll use a simple quadratic function for this exercise: f(x)=x^2

### Gradient Descent Steps
1. Start with an initial guess for x. 
2. Compute the gradient (derivative) of f at this point.
3. Update x by moving in the direction opposite to the gradient.
4. Repeat until convergence.


# Setup
Import necessary libraries and define the function and its derivative.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x ** 2

def df(x):
    return 2 * x
```

# Simulation
Manually perform gradient descent on the function f(x).

```python
def gd(learning_rate_modifier = 1, starting_point_modifier = 1, plot_title=""):
    # Gradient Descent Parameters
    x = 10 * starting_point_modifier  # initial guess
    learning_rate = 0.1 * learning_rate_modifier
    iterations = 25

    # Lists to store x values and f(x) values
    x_values, f_values = [], []

    # Gradient Descent Loop
    for i in range(iterations):
        x_values.append(x)
        f_values.append(f(x))

        # Calculate the gradient
        grad = df(x)

        # Update x
        x = x - learning_rate * grad

    # Plotting the function and the steps
    plt.plot(np.linspace(-10, 10, 100), f(np.linspace(-10, 10, 100)), label='f(x) = x^2')
    plt.scatter(x_values, f_values, color='red')
    plt.title('Gradient Descent on f(x) = x^2' + plot_title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

gd()
```

# Tasks and Questions
1. Run the Simulation: Execute the above code and observe the red dots converging to the minimum of the function.
2. Adjust Parameters: Try different learning rates and starting points. Observe how these changes affect the convergence.
3. Answer Questions:

```python
gd(learning_rate_modifier=0.1, plot_title=" (learning_rate = 0.1)")
gd(learning_rate_modifier=8, plot_title=" (learning_rate = 8)")
gd(starting_point_modifier=0.25, plot_title=" starting_point = 5")
gd(starting_point_modifier=2, plot_title=" starting_point = 20")
```
